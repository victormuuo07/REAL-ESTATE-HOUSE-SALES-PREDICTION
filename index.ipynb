Distribution of Key Variables:   

Purpose:  

Understand the distribution of your key variables such as Median_House_Value, Median_Income, and Population.

```python
df_2['Median_House_Value'].hist(bins=50)
plt.title('Distribution of House Values')
plt.xlabel('House Value')
plt.ylabel('Frequency')
plt.show()
```

From the histogram you generated, here's what we can infer about the distribution of house values:     

Skewed Distribution:   

The data appears to be right-skewed or positively skewed. This means that while most house values cluster around the lower range (closer to 0), there are a few high-value houses that create a long tail to the right. This indicates that the majority of houses are relatively affordable, but there are some very expensive houses that are pulling the average up.   


High Frequency at Lower Values:  

There is a high frequency of houses in the lower range of the value spectrum (below 200,000). This suggests that a significant portion of homes in the dataset are priced relatively affordably, which could indicate a large portion of the population is living in these homes.   

Outliers:  

There is a sharp spike at the 500,000 mark, which could indicate that there are outliers or extremely high-value properties. This may distort the overall analysis and skew any models or averages you may compute later. It's worth checking whether these extreme values are valid data points or if they need to be handled (e.g., removed or capped).  

Potential for Log Transformation:   

Given the right skew, a log transformation may be appropriate if you're going to use machine learning models. This can help normalize the distribution, making the model assumptions more valid.  

Distribution Shape:   

The distribution of house values is common in real estate datasets, where most homes are priced within a reasonable range, and only a few high-end properties stand out.  

**Capping the Outliers**

```python
import numpy as np
# Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'Median_House_Value'
Q1 = df_2['Median_House_Value'].quantile(0.25)
Q3 = df_2['Median_House_Value'].quantile(0.75)

# Step 2: Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Step 3: Define lower and upper bounds to detect outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Step 4: Check for outliers
outliers = df_2[(df_2['Median_House_Value'] < lower_bound) | (df_2['Median_House_Value'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

# Step 5: Cap the outliers (optional step)
df_2['Median_House_Value'] = np.where(df_2['Median_House_Value'] > upper_bound, upper_bound,
                                       np.where(df_2['Median_House_Value'] < lower_bound, lower_bound, df_2['Median_House_Value']))

#  Check the distribution again after capping


df_2['Median_House_Value'].hist(bins=50)
plt.title('Distribution of House Values after Capping Outliers')
plt.xlabel('House Value')
plt.ylabel('Frequency')
plt.show()
```
