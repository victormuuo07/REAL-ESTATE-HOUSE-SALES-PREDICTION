```python
import pandas as pd
import os

# Define the directory containing the datasets
data_dir = '/kaggle/input/property-sales-data-real-estate-trends/'

# List of files you want to load and combine
files = [
    '2002-property-sales-data.csv',
    '2003-property-sales-data.csv',
    '2004-property-sales-data.csv',
    '2005-property-sales-data.csv',
    '2006-property-sales-data.csv',
    '2007-property-sales-data.csv',
    '2008-property-sales-data.csv',
    '2009-property-sales-data.csv',
    '2010-property-sales-data.csv',
    '2011-property-sales-data.csv',
    '2012-property-sales-data.csv',
    '2013-property-sales-data.csv',
    '2014-property-sales-data.csv',
    '2015-property-sales-data.csv',
    '2016-property-sales-data.csv',
    '2017-property-sales-data.csv',
    '2018-property-sales-data.csv',
    '2019-property-sales-data.csv',
    '2020-property-sales-data.csv',
    '2021-property-sales-data.csv',
    '2022-property-sales-data.csv'
]

# Initialize an empty list to store dataframes
df_list = []

# Loop through the files and process each one
for file in files:
    # Read the CSV file
    file_path = os.path.join(data_dir, file)
    df = pd.read_csv(file_path)
    
    # Extract the year from the filename (e.g., '2002' from '2002-property-sales-data.csv')
    year = file.split('-')[0]
    
    # Add a Year column
    df['Year'] = int(year)
    
    # Append the dataframe to the list
    df_list.append(df)

# Concatenate all the dataframes into a single dataframe
combined_df = pd.concat(df_list, ignore_index=True)

# Display the first few rows of the combined dataframe
combined_df.head()
```
